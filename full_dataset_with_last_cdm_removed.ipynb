```python
import pandas as pd

df = pd.read_csv('train_data.csv')
```

```python
df_filtered = df[df['time_to_tca'] > .5]
```

```python
display(df_filtered)
print(df_filtered.shape)
```

```python
df_filtered_no_na = df_filtered.dropna()
print(df_filtered_no_na.shape)
```

```python
df_filtered_relevant_columns = df_filtered_no_na[['event_id', 'risk', 'max_risk_scaling', 'max_risk_estimate', 'time_to_tca', 'mahalanobis_distance', 'miss_distance', 'relative_position_r', 'relative_position_t', 'relative_position_n', 'relative_velocity_r',
'relative_velocity_t', 'relative_velocity_n', 'c_weighted_rms', 't_weighted_rms', 'c_sigma_tdot', 't_span', 't_j2k_inc', 'c_j2k_inc', 'c_sigma_r',
'c_crdot_t', 'c_sigma_n', 'c_sigma_t', 'c_sigma_r', 'c_ctdot_t', 'c_position_covariance_det', 't_position_covariance_det']]
display(df_filtered_relevant_columns)
```

```python
grouped_by_event = df_filtered_relevant_columns.groupby('event_id')
print(grouped_by_event.ngroups)
```

Each group is one event. There are a different number of CDMs in each event. You can access each event by calling grouped_by_event.get_group(x). The 25 features to be used have already been accounted for and the others have been removed. The loop below will help to access the features and target columnn for each event and store them for future use.

```python
target_column = 'risk'

# Initialize lists to store features and targets
all_features = []
all_targets = []

for event_id, group in grouped_by_event:
    features = group.drop(columns=[target_column]).values  # Convert to numpy array (optional)
    target = group[target_column].values  # Convert to numpy array (optional)

    all_features.append(features)
    all_targets.append(target)
```
